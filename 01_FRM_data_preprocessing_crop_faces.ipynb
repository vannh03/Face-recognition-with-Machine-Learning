### Data Preprocessing
**1. Detect Face and Crop from Image:**

In this notebook you will learn how to crop faces from all images in male and female folder

- step-1: Read Image and convert to RGB
- step-2: Apply haar cascade classifier
- step-3: Crop face using coordinates from cascade classifier
- step-4: save image in repective directory/folder

```python
import numpy as np # numerical python
import cv2 # opencv
from glob import glob # glob use to extract path of file
import matplotlib.pyplot as plt # visualze and display 
```

1. **Get list of all images path**

```python
hpath = glob('./data/Henry/*.jpg') # this will return all images path in a list
vpath = glob('./data/Van/*.jpg')
```

```python
print('The number of images in Henry folder = ',len(hpath))
print('The number of images in Van folder = ',len(vpath))
```

2. **Understand the 4 steps**

```python
### Step -1 Read Image and Convert to RGB
img = cv2.imread(hpath[0]) # read image in BGR
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # this step will convert image from BGR to RGB

### Step -2: Apply Haar Cascade Classifier
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
faces_list = haar.detectMultiScale(gray,1.5,5)
print(faces_list)
for x,y,w,h in faces_list:
    cv2.rectangle(img_rgb,(x,y),(x+w,y+h),(0,255,0),2)
    ### Step -3: crop Face 
    roi = img_rgb[y:y+h,x:x+w]
    
    plt.imshow(roi)
    plt.axis('off')
    plt.show()
    

### Step -4: Save Image
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

3. Implement above process to all images

**Crop all `Henry` Images**

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
```

```python
for i in range(len(hpath)):
    try:
            
        ### Step -1 Read Image and Convert to RGB
        img = cv2.imread(hpath[i]) # read image in BGR
        img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # this step will convert image from BGR to RGB

        ### Step -2: Apply Haar Cascade Classifier
        gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
        faces_list = haar.detectMultiScale(gray,1.5,5)
        for x,y,w,h in faces_list:
            ### Step -3: crop Face 
            roi = img[y:y+h,x:x+w]
            ### Step -4: Save Image
            cv2.imwrite(f'./crop_data/Henry/Henry_{i}.jpg',roi)
            print('Image Sucessfully processed')
    except:
        print('Unable to Process the image')
```

**Crop all `Van` Images**

```python
for i in range(len(vpath)):
    try:
            
        ### Step -1 Read Image and Convert to RGB
        img = cv2.imread(vpath[i]) # read image in BGR
        img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # this step will convert image from BGR to RGB

        ### Step -2: Apply Haar Cascade Classifier
        gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
        faces_list = haar.detectMultiScale(gray,1.5,5)
        for x,y,w,h in faces_list:
            ### Step -3: crop Face 
            roi = img[y:y+h,x:x+w]
            ### Step -4: Save Image
            cv2.imwrite(f'D:/Assigment_CPV/crop_data/Van/Van_{i}.jpg',roi)
            print('Image Sucessfully processed')
    except:
        print('Unable to Process the image')
```
